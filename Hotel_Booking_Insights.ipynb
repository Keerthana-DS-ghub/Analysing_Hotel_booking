**Printing revenue realized per hotel type**

```python
revenue_per_hotel = df_bookings_all.groupby("category")["revenue_generated"].sum()
revenue_per_hotel
```

**Visualizing revenue per hotel type, use Matplotlib**

```python
revenue_per_hotel.plot(kind='bar', color='skyblue', figsize=(8,5))
plt.xlabel("Hotel Type")
plt.ylabel("Total Revenue")
plt.title("Revenue Realized per Hotel Type")
plt.xticks(rotation=45)
plt.show()
```

**Print a pie chart of revenue realized per booking platform bold text**

```python
revenue_per_platform = df_bookings_all.groupby("booking_platform")["revenue_generated"].sum()
# Plot Pie Chart
plt.figure(figsize=(7, 7))
revenue_per_platform.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=["skyblue", "lightgreen", "coral", "gold","blue"])
plt.title("Revenue Distribution Per Booking Platform")
plt.ylabel("")  # Hide y-label for better visualization
plt.show()
```
